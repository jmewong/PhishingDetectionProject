```julia
function produce_proportions(col_number, y_vector, dataset)
    l = length(y_vector)
    #number of classified "phishing" for each category
    count_negative_one = 0
    count_zero = 0
    count_positive_one = 0
    for i=1:l
        if y_vector[i] == 1
            if dataset[i,col_number] == -1
                count_negative_one += 1
            end

            if dataset[i,col_number] == 0
                count_zero += 1
            end

            if dataset[i,col_number] == 1
                count_positive_one += 1
            end
        end
    end
    
    a = count_negative_one / l, count_zero / l, count_positive_one / l
    return [x for x in a]
end

function prop_compare(col_number, y_vector, dataset)
    x = produce_proportions(col_number, y_vector, dataset)
    y = produce_proportions(col_number, dataset[:,:Result], dataset)
    
    return [x y]
end
prop = prop_compare(1, ylog, dataset)

#plot colorcoded heatmap of binary values (% classified as phishing or legitimate)
#compare to the true heatmap
#do it for each feature
#do it for each model
```

```julia
for i in 1:size(dataset)[2]
    PyPlot.figure()
    fig = imshow(prop_compare(i, ylog, dataset), cmap="hot", interpolation="bilinear", 
        origin="upper", extent=(0,1,-1,1))
    suptitle("Logistic Regression: " * String(DataFrames.names(dataset)[i]))
    locs, _ = xticks()
    xticks([.2,.8], ["true", "predicted"])
    locs, _ = yticks()
    yticks(locs, ["-1", "","","","0","","","", "1"])
    colorbar()
    fig
end
```
